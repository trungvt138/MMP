```python
def perceptron_learning(x_data, y_data, learning_rate=0.1, epochs=100):
    num_features = x_data.shape[1]
    num_samples = x_data.shape[0]

    # Initialize weights and bias
    weights = np.random.randn(num_features)
    bias = np.random.randn()

    accuracy_list = []

    for epoch in range(epochs):
        correct_predictions = 0

        for i in range(num_samples):
            # Forward pass
            predicted_output = np.dot(x_data[i], weights) + bias
            predicted_label = 1 if predicted_output >= 0 else 0

            # Update weights and bias
            error = y_data[i] - predicted_label
            weights += learning_rate * error * x_data[i]
            bias += learning_rate * error

            # Check if prediction is correct
            if predicted_label == y_data[i]:
                correct_predictions += 1

        # Calculate accuracy for the epoch
        accuracy = correct_predictions / num_samples
        accuracy_list.append(accuracy)

    return weights, bias, accuracy_list

# Example usage

weights, bias, accuracy_list = perceptron_learning(x_data, y_data)
print("Weights:", weights)
print("Bias:", bias)

# Plot accuracy
plt.plot(accuracy_list)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy of Perceptron Learning Algorithm')
plt.show()
```
